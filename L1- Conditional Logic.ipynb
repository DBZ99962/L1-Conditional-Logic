### 1.

Go through the string below and if the length of a word is even print "even!"

string = 'I know how to print every word in this sentence that has an even number of letters'

```python
# Given string
string = 'I know how to print every word in this sentence that has an even number of letters'

# Split the string into words
words = string.split()

# Iterate through each word
for word in words:
    # Check if the length of the word is even
    if len(word) % 2 == 0:
        # Print the word if its length is even
        print('even')
```

### 2. 

Write a program to check given number is odd or even

Hint:
- 20 is even
- 27 is odd

```python
# Define a number to check
num = int(input("Enter a number: "))

# Check if the number is even using modulo operator
if num % 2 == 0:
    # Print "even" if remainder is 0
    print("even")
else:
    # Print "odd" if remainder is 1
    print("odd")
```

### 3.

Write a program to check given number is divisible by 6 or not

Hint:
24 is divisible by 6
27 is not divisible by 6

```python
# Define a number to check
num = int(input("Enter a number: "))

# Check if the number is divisible by 6 using modulo operator
if num % 6 == 0:
    # Print "divisible by 6" if remainder is 0
    print("divisible by 6")
else:
    # Print "not divisible by 6" if remainder is not 0
    print("not divisible by 6")
```

Create a list containing the following 3 elements:

- Subject person had taken

print all details and also print person has programming experience or not

```python
# Create a list with subject details
subject_name = "Python"
marks_achieved = 95
has_programming_experience = True

# Create list with 3 elements
person_details = [subject_name, marks_achieved, has_programming_experience]

# Print all details
print("Subject:", person_details[0])
print("Marks Achieved:", person_details[1])
print("Programming Experience:", person_details[2])

# Check and print if person has programming experience
if person_details[2]:
    # Print message if person has programming experience
    print("Person has programming experience")
else:
    # Print message if person does not have programming experience
    print("Person does not have programming experience")
```

### 5.

Write a program to check sum of all elements in a list is greater than 100 or not

Hint:
- [11, 13, 17, 19, 11] --> sum is not greater than 100
- [24, 36, 30, 31, 42] --> sum is greater than 100

```python
# Define lists to test
list1 = [11, 13, 17, 19, 11]
list2 = [24, 36, 30, 31, 42]

# Check if sum of list1 is greater than 100
sum1 = sum(list1)
if sum1 > 100:
    # Print message if sum is greater than 100
    print(f"Sum {sum1} is greater than 100")
else:
    # Print message if sum is not greater than 100
    print(f"Sum {sum1} is not greater than 100")

# Check if sum of list2 is greater than 100
sum2 = sum(list2)
if sum2 > 100:
    # Print message if sum is greater than 100
    print(f"Sum {sum2} is greater than 100")
else:
    # Print message if sum is not greater than 100
    print(f"Sum {sum2} is not greater than 100")
```
